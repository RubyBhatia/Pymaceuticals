# Pymaceuticals Inc.

The main objective of this analysis was to compare the performace of Pymaceuticals drugs named Capomulin with other treatment regimens. To evaluate 250 mice indentifies with SSC tumor growth. These were treated through a variety of drug regimens over a course of 45 days where the main observation and mesurmented was based on the development of tumor. For the purpose of analysis, four promising treatment regimens were take such as Capomulin, ramicane, Infubinol and Ceftamin. 

Looking at the data, it can be seen that:
1. Capomulin had the hightest tumor volume of 230 with a Median of 41.56
2. Ramicane had 228 with a Median of 40.68
3. Infubinol had 178 with a Median of 51.82
4. Ceftamin had 178 with a Median of 51.78

These data explored that statistical significance since is obtained with a sample size =>30 as the number of data points for each drug is above 100. 

As per the data related to the average final tumor volume, it seems like Capomulin and Ramicane both were very powerful in minimusing the risk of tumor volume in the mice over the 45 day study period compare to Infubinol and Ceftamin. Capomuline is more cosistent in its effects than Ramicane as per the outcome of the shortness of its box. 

The line ploot 'Timepoint Versus Tumore Volume' demonstrates a downward trend of tumor shrinkage in the mice treated with the Capomulin treatment regimen. This suggests that prolonged and consistant treatment with Capomulin may give possitive result in reducing the tumor size, assuming other variables remain constant. 

There was a strong correlation(R-sqared of 0.84) between the weight of the mice and thier average tumor volume. This shows that mouse weight must be controlled to obtain accurate results. If only the final tumor volume is considered without accounting for mouse weight, it could lead to a skewed interpretation of the drug's efficacy.

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_reg_data = clean_total_data.set_index(["Drug Regimen"])
group_drug_data = drug_reg_data.groupby("Drug Regimen")
drug_reg_mean = group_drug_data["Tumor Volume (mm3)"].mean()
drug_reg_median = group_drug_data["Tumor Volume (mm3)"].median()
drug_reg_var = group_drug_data["Tumor Volume (mm3)"].var()
drug_reg_std = group_drug_data["Tumor Volume (mm3)"].std()
drug_reg_sem = group_drug_data["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
drug_reg_sum_df = pd.DataFrame({"Mean Tumore Volume": drug_reg_mean,
                                "Median Tumore Volume": drug_reg_median,
                                "Tumore Volume Variance": drug_reg_var,
                                "Tumor Volume Std. Dev.": drug_reg_std,
                                "Tumore Volume Err.": drug_reg_sem,})
drug_reg_sum_df                               
                                
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_drug_reg_sum_df = drug_reg_data.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean","median","var","std","sem"])
agg_drug_reg_sum_df
```

```python
import pandas as pd



df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

#Put treatments into a list for the loop (and later for plot labesls)
regimenz = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create an empty list to fill with tumor volume data (for plotting)
final_vol = []

# calculate the IQR and quantitatively determine if there are any potential outliers. 

for regimen in regimenz:
    
    # Filter data for the current regimen
    tumor_vol_data = df.loc[df['Drug Regimen'] == regimen, 'Tumor Volume (mm3)']

    # Append tumor volume data to the final_vol list
    final_vol.append(tumor_vol_data)    

    # Calculate quartiles, IQR, and potential outliers
    quartiles = tumor_vol_data.quantile([0.25, 0.75])
    Q1 = quartiles[0.25]
    Q3 = quartiles[0.75]
    Iqr = Q3 - Q1
    
    # Determine outliers using the IQR method
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = tumor_vol_data.loc[(tumor_vol_data < lower_bound) | (tumor_vol_data > upper_bound)]

    # Print results
    print(f"results for {regimen}:")
    print(f" - Quartile 1 (Q1): {Q1}")
    print(f" - Quartile 1 (Q1): {Q3}")
    print(f" - Interquartile range (IQR): {IQR}")
    print(f" - Values below {lower_bound} or above {upper_bound} could be outliers.")

    if not outliers.empty:
        print(f" - Potential outliers: {outliers.values}")
    else:
        print(" - No potential outliers.")
    print()
   
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# seed the random number generator for reproducibility
np.random.seed(0)

# Creat a smaple DataFrame with Treatment and Tumor Volume data
data = {
    'Treatment': np.random.choice(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'], size=100),
    'Tumor Volume (mm3)': np.random.normal(loc=50, scale=10, size=100)
}
sample_df = pd.DataFrame(data)

# Generate a line plot of tumor volume vs. time point
plt.figure(figsize=(8, 4))
sns.boxplot(x='Treatment', y='Tumor Volume (mm3)', data=sample_df)
plt.title('Tumor Volume Distribution by Treatment Group')
plt.xlabel('Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Select a single mouse treated with Capomulin

capomulin_mouse = df[(df['Drug Regimen'] == 'Capomulin') & (df['Mouse ID'] == df[df['Drug Regimen'] == 'Capomulin']['Mouse ID'].iloc[0])]
# Generate a line plot of tumor volume vs. time point
plt.figure(figsize=(8, 4))
plt.plot(capomulin_mouse['Timepoint'], capomulin_mouse['Tumor Volume (mm3)'], marker='o')
plt.title('Tumor Volume vs. Time Point for a Singe Mouse Treated with Capomulin')
plt.xlabel('Time Point')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Calculate the average observed tumor volume for each mouse in the Capomulin regimen
capomulin_data = df[df['Drug Regimen'] == 'Capomulin']
capomulin_avg_tumor_volume = capomulin_data.groupby('Mouse ID').agg({'Tumor Volume (mm3)': 'mean', 'Weight (g)': 'first'}).reset_index()

# Generate a scatter plot of mouse weight vs. the average observed tumor volume
plt.figure(figsize=(8, 4))
plt.scatter(capomulin_avg_tumor_volume['Weight (g)'], capomulin_avg_tumor_volume['Tumor Volume (mm3)'], marker='o')
plt.title('Mouse Weight vs. Average Tumor Volime for capomulin Regimen')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```
